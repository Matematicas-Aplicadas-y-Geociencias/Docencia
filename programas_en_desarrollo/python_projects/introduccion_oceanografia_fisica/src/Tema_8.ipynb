```python
######################################################
# Formula de Eckart (1952) para calcular longitud de onda
# aproximada con un 10% de error
######################################################

# valores constantes
g = 9.81    # gravity

# variables conocidas
T = np.arange(3,22,3)           # period
h = np.arange(0, 201) # =20     # %depth, this can be an array 

# ecuacion principal
LEk = np.zeros((h.size, T.size))
for i in range(T.size):
    LEk[:, i] = (g * (T[i]**2) / (2 * PI)) * np.sqrt(np.tanh(((4 * PI**2) * h) / (g * (T[i]**2))))

plt.figure(figsize=(10, 4))
plt.plot(h,LEk)
plt.legend(['3','6','9','12','15','18','21'])
plt.title('Longitudes de onda para T variable')
plt.xlabel('profundidad (m)')
plt.ylabel('longitud de onda')
```

```python
######################################################
######################################################
# Iteracin para calcular L en cualquier profundifdad relativa
n = 100          # number of iterations
# Valor semilla de longitud de onda (valor en aguas profundas)
T2 = 15
h2 = 30
L1 = 1.56 * T2**2
L3 = np.zeros(n)
L4 = np.zeros(n)

for k in range(n):
    L3[k]   = L1
    L2      = (g * (T2**2) * 0.5 / PI) * np.tanh(2 * PI * h2 / L1)   # check if it's right
    L1      = L2  # redo
    L4[k]   = L2

plt.plot(L3,'.-', linewidth=1)
plt.plot(L4,'.-r', lw=1)
plt.legend(['L1','L2'])
```
